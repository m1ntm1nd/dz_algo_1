#1. Найдите значение функции в точках 1, 5, 10, 100

$$ y = \log_{2x}\frac{e^{cos(x)^A}tan(x^2)}{\frac{A}{9}x +\sqrt{x^3}} $$
*A*  = 24

```python
import numpy as np
CONST_A = 24

values = [1, 5, 10, 100]

numen = np.exp(np.power(np.cos(values), CONST_A, dtype = np.complex)) * np.tan(np.divide(np.power(values, 2),np.pi/180))
denom = np.multiply(values, CONST_A/9) + np.sqrt(np.power(values, 3))
func = np.divide(numen, denom)
print(np.log(func) / np.log(np.multiply(values,2)))
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open('lab1ex2.csv','r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)

    data = {} #represents data in view {w:[Ur,Uc,Ul]}
    for line in csv_reader:
        keyss = float(line[0].replace(',','.'))
        valuess = float(line[1].replace(',','.'))
        data[keyss] = valuess
```

```python
    x = list(data.keys())
    y = list(data.values())
    plt.figure(figsize=(10, 5))
    plt.plot(x,y,label = r"Ur")
    plt.errorbar(x, y, xerr=0.05, yerr=2,marker='s', mfc='red', mec='green')
    x2 = np.arange(1.0,16.0,0.01)
    y2 = np.divide(200,x2)
    plt.plot(np.add(x2,3),y2,label = r"$200/x-4$")
    plt.xlabel('w, kHz', fontsize=16)
    plt.ylabel('U, mV', fontsize=16) 
    plt.grid(True)
    plt.legend(loc='best', fontsize=12)
    plt.savefig('figure_with_legend.png')
    plt.show()
```
