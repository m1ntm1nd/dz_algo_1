```python
eps = 10 ** -3
```

```python
c = 1 - 2/np.sqrt(np.pi)*eps
k = 2
```

```python
def func(x):
    return (np.exp(-x**2)*np.sin(x))
#print(func(1))
def integral(x):
    s = 0
    for i in range(len(x)-1):
        xi = (x[i+1]+x[i])/2
        s = s + func(xi)*(x[i+1]-x[i])
    return s

def f38(x):
    s = 0
    for i in range(len(x)-1):
        xi = (x[i+1]-x[i])
        s = s + (xi*(func(x[i]) + 3*func((2*x[i]+x[i+1])/3) + 3*func((x[i]+2*x[i+1])/3) + func(x[i+1]))/8)
    return s
```

```python
n = 2
x1 = np.linspace(1,2,n)
#print(x1)
#print(integral(x1),n)
x2 = np.linspace(1,2,5)
while((integral(x2)-integral(x1))>eps):
    x1 = x2
    n+=1
    x2 = np.linspace(1,2,n)
print(integral(x2),n)
```

```python
n = 2
x1 = np.linspace(1,2,n)
#print(x1)
#print(integral(x1),n)
x2 = np.linspace(1,2,5)
while((f38(x2)-f38(x1))>eps):
    x1 = x2
    n+=1
    x2 = np.linspace(1,2,n)
print(f38(x2),n)
```
