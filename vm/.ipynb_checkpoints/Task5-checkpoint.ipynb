```python
fid=open('data_var07.txt',"r")
lines=fid.readlines()
fid.close()
lines=lines[1:]
Fullo=[]
Fullh=[]
Fulll=[]
Fullc=[]
Fullb=[]
n=1
for i in range(len(lines)): 
    data=float(lines[i].split(',')[1])
    Fullo.append(data)
    data1=float(lines[i].split(',')[2])
    Fullh.append(data1)
    data2=float(lines[i].split(',')[3])
    Fulll.append(data2)
    data3=float(lines[i].split(',')[4])
    Fullc.append(data3)
    
while n<=(len(lines)):
    data4=float(n)
    Fullb.append(data4)
    n=n+1
    
print (Fullb)
print (Fullc)
```

```python
#easyappr
import scipy as sp
import matplotlib.pyplot as plt
fx = sp.linspace(Fullb[0], Fullb[-1]+1, 1000)
for d in range (1, 6):
    fp, residuals, rank, sv, rcond = sp.polyfit(Fullb, Fullo, d, full=True)
    f=sp.poly1d(fp)
    plt.plot (Fullb, Fullo, 'o', markersize=1)
    plt.plot (fx, f(fx), linewidth=4)
    plt.show()
```

```python
#hardappr
import numpy as np
from numpy import linalg
import pylab
from matplotlib import mlab
p=5
A=np.zeros((p+1,p+1))
g=[0 for i in range(len(lines))]
for i in range(p+1):
    for j in range(p+1):
        for k in range(len(lines)):
            g[k]=k**(2*p-j-i)
        A[i,j]=sum(g)
B=np.zeros((p+1,1))
h=[0 for i in range(len(lines))]
for l in range(p+1):
    for r in range(len(lines)):
        h[r]=Fullo[r]*r**(p-l)
    B[l,0]=sum(h)
C=linalg.solve(A,B)
def fun(x):
    z=0
    for s in range(p+1):
        z=z+C[s]*x**(p-s)
    return z

xlist=mlab.frange(-0.5,3200.5,0.1)
ylist=[fun(x) for x in xlist]
plt.grid (True)
plt.plot (Fullb, Fullo, 'o', markersize=3)
plt.plot (xlist,ylist, linewidth=4)
plt.show()
print (A)
print (B)
print (C)
```
